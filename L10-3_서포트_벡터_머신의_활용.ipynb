## **회귀 모델 사례: 급여 데이터**

```python
import pandas as pd

salary = pd.read_csv('Salary_Data.csv')
salary
```

```python
salary.info()
```

```python
salary = salary.dropna()
salary
```

```python
set(salary['Gender'])
```

```python
len(salary[salary['Gender'] == 'Other'])
```

```python
salary = salary.drop(salary[salary['Gender'] == 'Other'].index)

Gender_mapping = {'Male': 0, 'Female': 1}
salary['Gender'] = salary['Gender'].replace(Gender_mapping)
salary
```

```python
set(salary['Education Level'])
```

```python
Education_mapping = {"Bachelor's": 1,
 "Bachelor's Degree": 1,
 'High School': 0,
 "Master's":2,
 "Master's Degree":2,
 'PhD':3,
 'phD':3}

salary['Education Level'] = salary['Education Level'].replace(Education_mapping)
salary
```

```python
len(set(salary['Job Title']))
```

```python
salary = pd.get_dummies(salary, columns = ['Job Title'])
salary
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(salary.drop(columns = 'Salary'), salary['Salary'],
                                                    test_size = 0.2, random_state=0)
```

```python
from sklearn.svm import SVR

model = SVR(kernel = 'linear', C = 1)
model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def adjusted_r2(r2, n, k):
  return 1 - (1-r2) * (n-1) / (n-k-1)

mae = mean_absolute_error(pred, y_test)
mse =mean_squared_error(pred, y_test)
r2 = r2_score(pred, y_test)
ad_r2 = adjusted_r2(r2, X_train.shape[0], X_train.shape[1])

print('MAE:', mae)
print('RMSE:', mse)
print('R2:', r2)
print('Adj R2:', ad_r2)
```

## **분류 모델 사례: 암진단 데이터**

```python
import pandas as pd

cancer = pd.read_csv('cancer_classification.csv')
cancer.head()
```

```python
diagnosis_mapping = {'B':0, 'M':1}
cancer['diagnosis'] = cancer['diagnosis'].replace(diagnosis_mapping)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.drop(columns= ['id', 'diagnosis']), cancer['diagnosis'],
                                                    test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', C = 1)
model.fit(X_train_scaled, y_train)

pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

test_cm = confusion_matrix(y_test, pred)
test_acc = accuracy_score(y_test, pred)
test_prc = precision_score(y_test, pred)
test_rcll = recall_score(y_test, pred)
test_f1 = f1_score(y_test, pred)

print(test_cm)
print()
print("정확도\t{}%". format(round(test_acc*100, 2)))
print("정밀도\t{}%". format(round(test_prc*100, 2)))
print("재현율\t{}%". format(round(test_rcll*100, 2)))
print("F1\t\t{}%". format(round(test_f1*100, 2)))
```

## **`연습문제` bodyPerformance.csv의 class 변수에 대해 서포트 벡터 머신을 사용하여 적절한 과정을 거쳐서 분류 모델을 훈련한 뒤 성능평가지표와 서포트 벡터의 대한 정보를 출력해보시오.**

```python
import pandas as pd

body = pd.read_csv('bodyPerformance.csv')
body
```

```python
gender_mapping = {'M':0, 'F':1}
body['gender'] = body['gender'].replace(gender_mapping)
body
```

```python
drop(columns=['컬럼명'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(body.drop(columns=['class']), body['class'], test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', C = 1)
model.fit(X_train_scaled, y_train)

pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

test_cm = confusion_matrix(y_test, pred)
test_acc = accuracy_score(y_test, pred)
test_prc = precision_score(y_test, pred, average='macro')
test_rcll = recall_score(y_test, pred, average='macro')
test_f1 = f1_score(y_test, pred, average='macro')

print(test_cm)
print()
print("정확도\t{}%". format(round(test_acc*100, 2)))
print("정밀도\t{}%". format(round(test_prc*100, 2)))
print("재현율\t{}%". format(round(test_rcll*100, 2)))
print("F1\t\t{}%". format(round(test_f1*100, 2)))
```
