### **이진 분류 모델 사례: 암진단 데이터**

```python
import pandas as pd

cancer = pd.read_csv('cancer_classification.csv')
cancer
```

```python
cancer.info()
```

```python
set(cancer['diagnosis'])
```

```python
diagnosis_mapping = {'B':0, 'M':1}
cancer['diagnosis'] = cancer['diagnosis'].replace(diagnosis_mapping)
```

```python
set(cancer['diagnosis'])
```

```python
cancer
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(cancer.drop(columns = ['id', 'diagnosis']), cancer['diagnosis'],
                                                    test_size = 0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)

x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

model_knn = KNeighborsClassifier()
model_logit = LogisticRegression(max_iter=500)

model_knn.fit(x_train_std, y_train)
model_logit.fit(x_train_std, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

pred_knn = model_knn.predict(x_test_std)

test_cm = confusion_matrix(y_test, pred_knn)
test_acc = accuracy_score(y_test, pred_knn)
test_prc = precision_score(y_test, pred_knn)
test_rcll = recall_score(y_test, pred_knn)
test_f1 = f1_score(y_test, pred_knn)

print('=== KNN ===')
print(test_cm, '\n')
print('정확도:', test_acc*100)
print('Precision:', test_prc*100)
print('Recall:', test_rcll*100)
print('F1-Score:', test_f1*100)

# ===== Logistic Regression =====
pred_logit = model_logit.predict(x_test_std)

logit_test_cm = confusion_matrix(y_test, pred_logit)
logit_test_acc = accuracy_score(y_test, pred_logit)
logit_test_prc = precision_score(y_test, pred_logit)
logit_test_rcll = recall_score(y_test, pred_logit)
logit_test_f1 = f1_score(y_test, pred_logit)

print('=== Logit ===')
print(logit_test_cm, '\n')
print('정확도:', logit_test_acc*100)
print('Precision:', logit_test_prc*100)
print('Recall:', logit_test_rcll*100)
print('F1-Score:', logit_test_f1*100)
```

### **이진 분류 모델 사례: 신체 능력 데이터**



```python
import pandas as pd

body = pd.read_csv('bodyPerformance.csv')
body
```

```python
body.info()
```

```python
print(set(body['gender']))
print(set(body['class']))
```

```python
gender_mapping = {'M':0, 'F':1}
class_mapping = {'A':0, 'B':1, 'C':2, 'D':3}

body['gender'] = body['gender'].replace(gender_mapping)
body['class'] = body['class'].replace(class_mapping)
```

```python
print(set(body['gender']))
print(set(body['class']))
```

```python
body.info()
```

```python
# 훈련 자료와 테스트 자료 분리 후, 스케일링
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(body.drop(columns = 'class'), body['class'], test_size = 0.2, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
# KNN과 로지스틱 회귀를 이용하여 다중분류 모델 생성

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

model_knn = KNeighborsClassifier()
model_knn.fit(x_train_std, y_train)

model_logit = LogisticRegression(multi_class='multinomial')
model_logit.fit(x_train_std, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

pred_knn = model_knn.predict(x_test_std)

test_cm = confusion_matrix(y_test, pred_knn)
test_acc = accuracy_score(y_test, pred_knn)
test_prc = precision_score(y_test, pred_knn, average='macro')
test_rcll = recall_score(y_test, pred_knn, average='macro')
test_f1 = f1_score(y_test, pred_knn, average='macro')

print('=== KNN ===')
print(test_cm)
print()
print('정확도:', test_acc*100)
print('Precision:', test_prc*100)
print('Recall:', test_rcll*100)
print('F1-Score:', test_f1*100)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

pred_logit = model_logit.predict(x_test_std)

test_cm = confusion_matrix(y_test, pred_logit)
test_acc = accuracy_score(y_test, pred_logit)
test_prc = precision_score(y_test, pred_logit, average='macro')
test_rcll = recall_score(y_test, pred_logit, average='macro')
test_f1 = f1_score(y_test, pred_logit, average='macro')

print('=== Logistic Regression ===')
print(test_cm)
print()
print('정확도:', test_acc*100)
print('Precision:', test_prc*100)
print('Recall:', test_rcll*100)
print('F1-Score:', test_f1*100)
```

### **연습 문제**
- `bodyPerformance.csv`의 `class`변수에 대해 결정 트리, 랜덤 포레스트, KNN, 로지스틱 회귀를 사용하여 적절한 과정을 거쳐서 분류 모델을 훈련한 뒤 각 모델의 성능을 비교해보시오.

```python
import pandas as pd

body = pd.read_csv('bodyPerformance.csv')

gender_mapping = {'M':0, 'F':1}
class_mapping = {'A':0, 'B':1, 'C':2, 'D':3}

body['gender'] = body['gender'].replace(gender_mapping)
body['class'] = body['class'].replace(class_mapping)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(body.drop(columns = 'class'), body['class'], test_size = 0.2, random_state=0)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)

x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model_tree = DecisionTreeClassifier()
model_tree.fit(x_train_std, y_train)

model_rf = RandomForestClassifier()
model_rf.fit(x_train_std, y_train)

model_knn = KNeighborsClassifier()
model_knn.fit(x_train_std, y_train)

model_logit = LogisticRegression(multi_class='multinomial')
model_logit.fit(x_train_std, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

pred_tree = model_tree.predict(x_test_std)
pred_rf = model_rf.predict(x_test_std)
pred_knn = model_knn.predict(x_test_std)
pred_logit = model_logit.predict(x_test_std)

tree_test_cm = confusion_matrix(y_test, pred_tree)
tree_test_acc = accuracy_score(y_test, pred_tree)
tree_test_prc = precision_score(y_test, pred_tree, average='macro')
tree_test_rcll = recall_score(y_test, pred_tree, average='macro')
tree_test_f1 = f1_score(y_test, pred_tree, average='macro')

rf_test_cm = confusion_matrix(y_test, pred_rf)
rf_test_acc = accuracy_score(y_test, pred_rf)
rf_test_prc = precision_score(y_test, pred_rf, average='macro')
rf_test_rcll = recall_score(y_test, pred_rf, average='macro')
rf_test_f1 = f1_score(y_test, pred_rf, average='macro')

knn_test_cm = confusion_matrix(y_test, pred_knn)
knn_test_acc = accuracy_score(y_test, pred_knn)
knn_test_prc = precision_score(y_test, pred_knn, average='macro')
knn_test_rcll = recall_score(y_test, pred_knn, average='macro')
knn_test_f1 = f1_score(y_test, pred_knn, average='macro')

logit_test_cm = confusion_matrix(y_test, pred_logit)
logit_test_acc = accuracy_score(y_test, pred_logit)
logit_test_prc = precision_score(y_test, pred_logit, average='macro')
logit_test_rcll = recall_score(y_test, pred_logit, average='macro')
logit_test_f1 = f1_score(y_test, pred_logit, average='macro')
```

```python
print('=== DecisionTreeClassifier ===')
print(tree_test_cm)
print()
print('정확도:', tree_test_acc*100)
print('Precision:', tree_test_prc*100)
print('Recall:', tree_test_rcll*100)
print('F1-Score:', tree_test_f1*100)
print()

print('=== RandomForestClassifier ===')
print(rf_test_cm)
print()
print('정확도:', rf_test_acc*100)
print('Precision:', rf_test_prc*100)
print('Recall:', rf_test_rcll*100)
print('F1-Score:', rf_test_f1*100)
print()

print('=== KNeighborsClassifier ===')
print(knn_test_cm)
print()
print('정확도:', knn_test_acc*100)
print('Precision:', knn_test_prc*100)
print('Recall:', knn_test_rcll*100)
print('F1-Score:', knn_test_f1*100)
print()

print('=== LogisticRegression ===')
print(logit_test_cm)
print()
print('정확도:', logit_test_acc*100)
print('Precision:', logit_test_prc*100)
print('Recall:', logit_test_rcll*100)
print('F1-Score:', logit_test_f1*100)
print()
```
