```python
# 필수 라이브러리
import numpy as np
import mglearn
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV, ElasticNetCV

# -----------------------------
# 1. 데이터 로드
# -----------------------------
X, y = mglearn.datasets.load_extended_boston()

# -----------------------------
# 2. 표준화 (필수)
# -----------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -----------------------------
# 3. alpha 후보 설정
# -----------------------------
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# -----------------------------
# 4. Lasso 회귀 (CV)
# -----------------------------
lasso_cv = LassoCV(
    alphas=alphas,
    cv=5,
    max_iter=100000
)

lasso_cv.fit(X_scaled, y)

print("===== Lasso 결과 =====")
print("최적 alpha:", lasso_cv.alpha_)
print("절편 (intercept):", lasso_cv.intercept_)
print("회귀 계수 (coef):")
print(lasso_cv.coef_)
print()

# -----------------------------
# 5. ElasticNet 회귀 (CV)
# -----------------------------
elastic_cv = ElasticNetCV(
    alphas=alphas,
    l1_ratio=0.5,   # Lasso : Ridge = 50 : 50
    cv=5,
    max_iter=100000
)

elastic_cv.fit(X_scaled, y)

print("===== ElasticNet 결과 =====")
print("최적 alpha:", elastic_cv.alpha_)
print("절편 (intercept):", elastic_cv.intercept_)
print("회귀 계수 (coef):")
print(elastic_cv.coef_)
```
