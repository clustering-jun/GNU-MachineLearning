```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

boston = pd.read_csv('boston.csv')

features = boston.drop(columns = 'PRICE')
target = boston['PRICE']

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = features.corr()
sns.heatmap(corr, annot = True, cmap='coolwarm', annot_kws={'size': 8}, fmt='.2f')
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(max_iter=1000, learning_rate='adaptive')
model.fit(features_scaled, target)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

boston = pd.read_csv('boston.csv')

features = boston.drop(columns = 'PRICE')
target = boston['PRICE']

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.5)
ridge.fit(features_scaled, target)
```

```python
ridge.intercept_
```

```python
ridge.coef_
```

```python
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas = [0.1, 1.0, 10.0])
ridge_cv.fit(features_scaled, target)
```

```python
ridge_cv.intercept_
```

```python
ridge_cv.coef_
```

```python
ridge_cv.alpha_
```

```python
# 

import numpy as np
import mglearn
from sklearn.linear_model import RidgeCV

# 1.  
X, y = mglearn.datasets.load_extended_boston()

# 2. alpha   
alphas = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

# 3. RidgeCV    
ridge_cv = RidgeCV(alphas=alphas)
ridge_cv.fit(X, y)

# 4.  alpha
best_alpha = ridge_cv.alpha_

# 5.   
intercept = ridge_cv.intercept_
coefficients = ridge_cv.coef_

# 6.  
print(" alpha:", best_alpha)
print(" (intercept):", intercept)
print("  (coefficients):")
print(coefficients)

```
