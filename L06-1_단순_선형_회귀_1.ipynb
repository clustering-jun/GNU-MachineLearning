```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('insurance.csv')
data
```

```python
x = data['age']
y = data['charges']

plt.figure(figsize=(10,5))
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
x = np.array(data['age'])
y = np.array(data['charges'])

x = x.reshape(1338, 1)
y = y.reshape(1338, 1)

model = LinearRegression()
model.fit(x, y)
```

```python
print('intercept:', model.intercept_, 'coef:', model.coef_)
```

```python
# íšŒê·€ ëª¨ë¸ ê²°ê³¼ì˜ ê²°ì •ê³„ìˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ìœ¼ë©° 4%ë¡œ ë‚®ê²Œ ë‚˜ì˜¨ ê²ƒì„ ì•Œ ìˆ˜ ìžˆìŒ
model.score(x, y)
```

```python
x_new = [[19], [64]]
y_hat = model.predict(x_new)
y_hat
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_new, y_hat, '-r')
plt.plot(x, y, 'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
