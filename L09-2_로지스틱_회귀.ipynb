### **로지스틱 회귀 모델 훈련하기**

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
features = iris.data[:100,:] # 행번호 0부터 99까지
target = iris.target[:100]

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)

model = LogisticRegression(random_state=0)
model.fit(features_standardized, target)
```

```python
obs = [[5, 5, 5, 5]]
print(model.predict(obs))
print(model.predict_proba(obs))
```

### **규제항을 사용한 이진분류**

```python
model_LR_pen = LogisticRegression(C=0.5, max_iter=200, penalty='l2', solver='sag', random_state=0)
model_LR_pen.fit(features_standardized, target)
```

```python
obs = [[5, 5, 5, 5]]
print(model_LR_pen.predict(obs))
print(model_LR_pen.predict_proba(obs))
```

### **다중 분류 로지스틱 회귀 모델 훈련하기**

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
features = iris.data
target = iris.target

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)


model_1 = LogisticRegression(random_state = 0, multi_class = "ovr")
model_1.fit(features_standardized, target)

observation = [[.5, .5, .5, .5]]
print(model_1.predict(observation))
print(model_1.predict_proba(observation))
```

```python
model_2 = LogisticRegression(random_state = 0, multi_class = "multinomial")
model_2.fit(features_standardized, target)

observation = [[.5, .5, .5, .5]]
print(model_2.predict(observation))
print(model_2.predict_proba(observation))
```

### **연습 문제**
- 아래의 코드를 참고하여 Survived에 대해 예측하는 로지스틱 회귀 모델을 적절하게 학습시킨 후, 다음 입력값에 대한 예측값과 예측 확률을 출력하시오.
- obs = [[1, 0, 28, 26]]


```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')
titanic = titanic.dropna()

features = titanic[['Pclass', 'Sex', 'Age', 'Fare']]
target = titanic['Survived']

Sex_mapping = {'male': 0, 'female':1}
features['Sex'] = features['Sex'].replace(Sex_mapping)

features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)

model_LR_pen_1 = LogisticRegression(C=0.5, max_iter=200, penalty='l2', solver='sag', random_state=0)
model_LR_pen_1.fit(features_standardized, target)
```

```python
obs = [[1, 0, 28, 26]]
print(model_LR_pen_1.predict(obs))
print(model_LR_pen_1.predict_proba(obs))
```

```python
obs = [[1, 0, 28, 26]]
obs_std = scaler.transform(obs)

print(model_LR_pen_1.predict(obs_std))
print(model_LR_pen_1.predict_proba(obs_std))
```
